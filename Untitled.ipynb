```python
import pandas as pd
import numpy as np

dates = pd.date_range(start='2024-01-01', periods=100)

sales = np.linspace(100, 500, 100) + np.random.randint(-20, 20, 100)

df = pd.DataFrame({
    'date': dates,
    'sales': sales.astype(int)
})

df.to_csv("sales.csv", index=False)

print("CSV file created!")
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Create 100 days
dates = pd.date_range(start='2024-01-01', periods=100)

# Generate features
marketing = np.random.randint(50, 200, 100)
discount = np.random.randint(5, 30, 100)

# Create realistic sales formula
sales = 50 + (marketing * 2) - (discount * 3) + np.random.randint(-20, 20, 100)

# Create dataframe
df = pd.DataFrame({
    'date': dates,
    'sales': sales,
    'marketing': marketing,
    'discount': discount
})

# Save CSV
df.to_csv("sales.csv", index=False)

print("✅ New CSV created with multiple features!")
```
